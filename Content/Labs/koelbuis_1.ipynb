# Theorie

Volgens [Newton's wet van afkoeling](https://nl.wikipedia.org/wiki/Afkoelingswet_van_Newton) is de snelheid waarmee een voorwerp afkoelt evenredig met het verschil in de temperatuur van het voorwerp ($T$) en de omgeving ($T_0$). We kunnen dit schrijven als:

$$
    \dot{Q} = -hA(T(t) - T_0),
$$

waarin 

- $\dot{Q}$ de warmtestroom in $\mathrm{W}$,
- $A$ het oppervlak waardoor koeling optreedt in $\mathrm{m}^2$,
- $h$ de warmteoverdrachtscoëfficiënt in $\mathrm{W/(m^2 K)}$.

Dit levert de differentiaalvergelijking

$$
    C\dot{T} = -hA(T(t) - T_0),
$$

met $C$ de warmtecapaciteit in $\mathrm{J/kg}$. Herschrijven met $\tau = \frac{C}{hA}$ levert:

$$
    -\tau\dot{T} = T(t) - T_0,
$$ (eq:dv1)

met als oplossing:

$$
    T(t) - T_0 = (T(0) - T_0)\text{e}^{-t/\tau}.
$$ (eq:dv1_opl)

We kunnen hieruit dus concluderen dat $\tau$ de karakteristieke tijdsduur is waarin de temperatuur van de buis een factor $\text{e}$ verlaagd ten opzichte van de omgevingstemperatuur.

Controleer dat vergelijking (4) een oplossing is van differentiaalvergelijking (3) door substitutie.

De gegeven differentiaalvergelijking is:
$$
-\tau \dot{T} = T(t) - T_0
$$

De voorgestelde oplossing luidt:
$$
T(t) - T_0 = \bigl(T(0) - T_0\bigr)e^{-t/\tau}
$$

# Stap 1: Bepaal de afgeleide van $T(t)$

Uit de oplossing volgt:
$$
T(t) = T_0 + \bigl(T(0) - T_0\bigr)e^{-t/\tau}
$$

De afgeleide is:
$$
\dot{T}(t) = -\frac{1}{\tau}\bigl(T(0) - T_0\bigr)e^{-t/\tau}
$$

# Stap 2: Substitueer in de differentiaalvergelijking

Linkerkant:
$$
-\tau \dot{T}(t) = -\tau \left(-\frac{1}{\tau}\bigl(T(0) - T_0\bigr)e^{-t/\tau}\right)
= \bigl(T(0) - T_0\bigr)e^{-t/\tau}
$$

Rechterkant:
$$
T(t) - T_0 = \bigl(T(0) - T_0\bigr)e^{-t/\tau}
$$

beide kanten zijn gelijk dus vergelijking 4 is een oplossing van differentiaalvergelijking 3


We zijn hier voor het gemak uitgegaan van een $h$ die onafhankelijk is van de temperatuur. We weten echter dat warmtetransport door straling niet lineair gaat, maar als 

$$
    \dot{Q}_s = \epsilon \sigma A (T^4 - T_0^4).
$$ 

Voor kleine temperatuurverschillen ($\Delta T = T - T_0$) is dit te vereenvoudigen tot

$$
    \dot{Q}_s = \epsilon \sigma A ((T_0+\Delta T)^4 - T_0^4) \approx \epsilon \sigma 4A T_0^3 \Delta T.
$$

Zolang $\Delta T$ dus relatief klein is ten opzichte van $T_0$, kunnen we $h$ dus inderdaad als een constante beschouwen. 

```{exercise} Uitwerken
:label: ex_uitw

Laat zien dat bovenstaande geldt door eerst $(T_0+\Delta T)^4$ uit te schrijven, te substitueren en dan te bedenken dat $\Delta T$ klein is en $\Delta T^2$ dus nog kleiner.
```

```{solution} ex_uitw```

We schrijven eerst $(T_0 + \Delta T)^4$ volledig uit met behulp van de binomiumformule:

$$
(T_0 + \Delta T)^4 = T_0^4 + 4T_0^3 \Delta T + 6T_0^2 \Delta T^2 + 4T_0 \Delta T^3 + \Delta T^4
$$

Substitutie in de uitdrukking voor het warmtetransport door straling geeft:

$$
\dot{Q}_s = \epsilon \sigma A \bigl((T_0 + \Delta T)^4 - T_0^4\bigr)
$$

$$
\dot{Q}_s = \epsilon \sigma A \left(4T_0^3 \Delta T + 6T_0^2 \Delta T^2 + 4T_0 \Delta T^3 + \Delta T^4 \right)
$$

Omdat $\Delta T$ klein is ten opzichte van $T_0$, zijn de termen met $\Delta T^2$, $\Delta T^3$ en $\Delta T^4$ veel kleiner dan de lineaire term en kunnen ze worden verwaarloosd. Dan blijft over:

$$
\dot{Q}_s \approx \epsilon \sigma 4A T_0^3 \Delta T
$$

Hieruit volgt dat het warmtetransport door straling voor kleine temperatuurverschillen lineair is in $\Delta T$.






```{exercise} 
:label: ex_fout
Hoe groot is de fout in het warmtetransport door straling die we maken voor het temperatuurbereik waarin we deze proef uitvoeren?
```

```{solution} ex_fout```

De fout die wordt gemaakt door het lineariseren van het warmtetransport door straling wordt bepaald door de verwaarloosde termen van orde $\Delta T^2$ en hoger.

De exacte uitdrukking is:
$$
\dot{Q}_{\text{exact}} = \epsilon \sigma A \left(4T_0^3 \Delta T + 6T_0^2 \Delta T^2 + 4T_0 \Delta T^3 + \Delta T^4 \right)
$$

De benaderde (lineaire) uitdrukking is:
$$
\dot{Q}_{\text{lin}} = \epsilon \sigma 4A T_0^3 \Delta T
$$

De relatieve fout is dus:
$$
\frac{\dot{Q}_{\text{exact}} - \dot{Q}_{\text{lin}}}{\dot{Q}_{\text{lin}}}
\approx \frac{6T_0^2 \Delta T^2}{4T_0^3 \Delta T}
= \frac{3}{2}\frac{\Delta T}{T_0}
$$

Voor een typisch experiment met $T_0 \approx 293\ \text{K}$ en een temperatuursverandering van bijvoorbeeld $\Delta T \leq 10\ \text{K}$ volgt:

$$
\frac{\text{fout}}{\text{signaal}} \approx \frac{3}{2}\frac{10}{293} \approx 0{,}05
$$

De fout door linearisatie bedraagt dus slechts enkele procenten binnen het temperatuurbereik van het experiment en is daarmee acceptabel.

Het buitenoppervlak van de buis wordt bepaald aan de hand van de geometrie. Voor een cilindrische buis met lengte $L$ en buitendiameter $D$ geldt (inclusief de twee uiteinden):

$$
A = \pi D L + 2 \cdot \pi \left(\frac{D}{2}\right)^2
$$

Indien één uiteinde is afgesloten met een dop en het andere niet bijdraagt aan warmte-uitwisseling, kan het oppervlak van één eindvlak eventueel worden weggelaten.

De warmtecapaciteit van de buis volgt uit:
$$
C_{\text{buis}} = m c
$$

waarbij de massa $m$ wordt berekend uit:
$$
m = \rho V
$$

en het volume $V$ van de buis:
$$
V = \pi \left(\frac{D_{\text{buiten}}^2 - D_{\text{binnen}}^2}{4}\right)L
$$

Hierin zijn $\rho$ en $c$ afhankelijk van het materiaal van de buis (zie tabel).

De omgevingstemperatuur $T_0$ is gelijk aan de kamertemperatuur en wordt bepaald door de thermometer af te lezen voordat de meting start, typisch rond $293\ \text{K}$.




Pak de buis op met thermisch isolerende handschoenen (of direct met de geïsoleerde grijparm) en plaats deze in de grijparm met isolatieschoentjes.
Positioneer de thermometer voor optimale temperatuurlezing.
Meet als functie van tijd hoe lichaam koelt.
Wacht voldoende lang zodat je de karakteristieke tijd $\tau$ voor de afkoeling kan bepalen.

De richting waarin de (cilindrische) as van de buis wordt gelegd, heeft invloed op het warmtetransport door convectie. 

Wanneer de buis verticaal is geplaatst, kan warme lucht langs het oppervlak gemakkelijker opstijgen, wat leidt tot sterkere natuurlijke convectie. In horizontale oriëntatie is deze luchtstroming minder efficiënt, waardoor de convectieve warmteoverdracht kleiner kan zijn. Dit heeft invloed op de gemeten karakteristieke afkoeltijd $\tau$.

De dop sluit één uiteinde van de buis af en vermindert daarmee het effectieve warmte-uitwisselingsoppervlak. Daarnaast voorkomt de dop dat lucht in de buis kan circuleren, wat anders extra warmteverlies door convectie aan de binnenzijde zou veroorzaken. Hierdoor is de afkoeling met dop langzamer dan zonder dop.



Doe dit voor twee of drie configuraties:

1. De buis met de as in verticale richting en afgesloten met dop.
2. De buis met de as in verticale richting *zonder* dop.
3. (alleen bij voldoende tijd) De buis met de as in horizontale richting en afgesloten met dop.

```{warning} Veiligheid
Zorg ervoor dat je de buis niet aanraakt terwijl deze warm is. De metalen buis geleidt warmte. Zelfs een temperatuur van 60 graden celsius is daarom al onaangenaam hoog. Hou de pan met het warme water ook op een veilige plek zodat deze niet kan omkiepen en het warme water over de grond opspat.
```

# Resultaten

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# afmetingen buis
lengte = 0.041          # m
diameter = 0.0195       # m
binnendiameter = 0.016  # m

# exponentiële afkoelfunctie
def exp_func(t, A, tau, T_omg):
    # A: beginverschil met omgeving
    # tau: karakteristieke afkoeltijd
    # T_omg: omgevingstemperatuur
    return A * np.exp(-t/tau) + T_omg  

# meetgegevens zonder dop
times1 = np.arange(0, 500, 10)
temps1 = np.array([
    44.1, 44.1, 44.1, 40.8, 40.8, 40.3, 40.3, 39.8, 39.2, 38.2,
    37.6, 37.6, 37.1, 36.6, 36.1, 35.6, 35.6, 35.2, 34.7, 34.2,
    33.7, 33.7, 33.2, 33.2, 32.7, 32.7, 32.3, 32.3, 32.3, 31.8,
    31.8, 31.3, 31.3, 30.8, 30.8, 30.3, 30.3, 30.3, 30.3, 30.3,
    29.9, 29.9, 29.9, 29.9, 29.4, 29.4, 29.4, 29.4, 29.0, 29.0
])

# meetgegevens met dop
times2 = np.arange(0, 500, 10)
temps2 = np.array([
    42.8, 42.8, 39.2, 38.7, 38.1, 37.6, 37.1, 36.4, 35.8, 35.3,
    34.9, 34.9, 34.5, 34.0, 34.0, 33.5, 33.5, 33.0, 32.9, 32.5,
    32.5, 32.3, 32.3, 31.8, 31.8, 31.7, 31.7, 31.7, 31.3, 31.3,
    31.3, 30.8, 30.8, 30.4, 30.4, 30.3, 30.3, 30.3, 30.0, 30.0,
    29.9, 29.8, 29.8, 29.8, 29.5, 29.5, 29.3, 29.2, 29.2, 29.0
])

# curve fit zonder dop
popt1, pcov1 = curve_fit(exp_func, times1, temps1,
                          p0=[22, 300, 22], maxfev=5000)

# curve fit met dop
popt2, pcov2 = curve_fit(exp_func, times2, temps2,
                          p0=[21, 200, 22], maxfev=5000)

# parameters uit fits
A_exp1, tau_exp1, T_omg_exp1 = popt1
A_exp2, tau_exp2, T_omg_exp2 = popt2

# berekende fits
y_fit1 = exp_func(times1, *popt1)
y_fit2 = exp_func(times2, *popt2)

# plot
plt.figure()
plt.xlabel('tijd (s)')
plt.ylabel('temperatuur (°C)')

plt.plot(times1, temps1, 'bo', label='metingen zonder stop')
plt.plot(times1, y_fit1, 'b-', label='$T = %0.2f e^{-t/%0.4f} + %0.2f$' % (A_exp1, tau_exp1, T_omg_exp1))

plt.plot(times2, temps2, 'ro', label='metingen met stop')
plt.plot(times2, y_fit2, 'r-', label='$T = %0.2f e^{-t/%0.4f} + %0.2f$' % (A_exp2, tau_exp2, T_omg_exp2))

plt.grid(True)
plt.legend()
plt.show()

# buitenoppervlak buis
buitenoppervlak = np.pi * lengte * diameter

# materiaalparameters (messing)
c = 380        # J/(kg K)
rho = 8730     # kg/m^3

# volume en massa
R = diameter / 2
r = binnendiameter / 2
V = np.pi * lengte * (R**2 - r**2)
m = rho * V

# warmtecapaciteit
warmtecapaciteit = m * c

# warmteoverdrachtscoëfficiënt
h_exp1 = warmtecapaciteit / (tau_exp1 * buitenoppervlak)
h_exp2 = warmtecapaciteit / (tau_exp2 * buitenoppervlak)

print(f"h zonder dop = {h_exp1:.2f} W/(m^2·K)")
print(f"h met dop   = {h_exp2:.2f} W/(m^2·K)")



````

# Discussie en conclusie

### Discussie

Uit de metingen blijkt dat de buis zonder dop langzamer afkoelt dan de buis met dop. Dit is terug te zien in de grotere karakteristieke tijd $\tau$ voor de configuratie zonder dop. De berekende warmteoverdrachtscoëfficiënt is zonder dop ongeveer 25 W/(m² K), terwijl deze met dop groter is, ongeveer 38 W/(m² K).

Het hogere $h$ bij de meting met dop kan worden verklaard door extra warmteverlies via het afgesloten uiteinde en mogelijk efficiëntere warmteoverdracht door convectie en straling via de dop. Daarnaast vormt de aanname van een constante warmteoverdrachtscoëfficiënt een vereenvoudiging. Ook warmtegeleiding via de grijparmen en meetonnauwkeurigheden van de thermometer kunnen bijdragen aan afwijkingen.

Ondanks deze beperkingen beschrijft het exponentiële model de temperatuurafname goed, wat aangeeft dat Newtons wet van afkoeling in dit temperatuurbereik een geschikte benadering is.

### Conclusie

De afkoeling van de metalen buis kan goed worden beschreven met een exponentiële functie zoals voorspeld door Newtons wet van afkoeling. Uit de metingen zijn karakteristieke afkoeltijden en bijbehorende warmteoverdrachtscoëfficiënten bepaald.

De warmteoverdrachtscoëfficiënt blijkt afhankelijk te zijn van de configuratie van de buis. In dit experiment is $h$ groter voor de buis met dop dan voor de buis zonder dop. Dit laat zien dat geometrie en randvoorwaarden een duidelijke invloed hebben op het warmtetransport, waarbij convectie en straling een belangrijke rol spelen.
